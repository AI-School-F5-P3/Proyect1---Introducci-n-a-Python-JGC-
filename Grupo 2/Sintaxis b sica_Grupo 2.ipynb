## Sintaxis básica
La sintaxis en lenguaje de programación hace referencia al conjunto de reglas que debe seguirse para para que el código pueda ser correctamente interpretado por humanos y máquinas.

La sintaxis de Python esta orientada a una alta legibilidad, de tal forma que la resolución de un código sea lo más obvio posible, prima la simplicidad. Para ello, la sintaxis emplea para sus palabras clave palabras en inglés. Esto genera que ciertas palabras estén reservadas por el lenguaje, tales como _while_, _for_, _repeat_...

Del mismo modo, Python reserva una serie de símbolos empleados para la definición de variables, funciones, argumentos, etc.

### Comentarios

La forma de mantener una comunicación interna en el código es mediante comentarios. Estas lineas de código tienen que estar especificadas de una forma concreta para evitar que se ejecuten al correr el programa. Para ello existen dos métodos:
- Hashtag ("#"). Define los comentarios en linea, para realizar aclaraciones cortas en el código.
- Comillas simples (''' '''), antes y después del texto. También conocido como un _docstring_ se emplea para explicaciones de mayor longitud, por ejemplo para definir los párametros de una función y como se ejecuta.

### Indentaciones

La indentación en Python se emplea como forma de _control flow_, es decir, delimitar que partes de código pertenecen a un mismo bloque. Por ejemplo, al definir una función, la indentación indica que lineas de código se incluyen en dicha función.

##**Operadores**

Los operadores en Python son símbolos especiales que realizan operaciones matemáticas, lógicas o de comparación en expresiones. Estos operadores se utilizan para realizar acciones como sumar, restar, multiplicar, dividir, comparar valores, asignar valores, entre otras.

En resumen, los operadores en Python son herramientas que nos permiten realizar diferentes operaciones matemáticas, lógicas o de comparación de forma sencilla y eficiente. Son fundamentales para manipular datos y controlar el flujo de un programa, facilitando la realización de tareas diversas de manera más rápida y efectiva.

### Operadores de Identidad


*   IS
*   IS NOT



```python
lista=[1,2,3,4]
lista2=['a','b','c']

n=2 #n='a'
if (n in lista) and (n in lista2):
    print(f'{n} está en ambas listas')

elif (n in lista) or (n in lista2):
    print(f'El dato {n} está en una sola lista.')
    if n in lista:
        print(f'El dato {n} está en la lista:{lista}')
        print(f'Entonces {n} es número?:{isinstance(n, int)}')
    elif n in lista2:
        print(f'La letra "{n}" está en la lista:{lista2}')
        print(f'Entonces {n} es letra?:{isinstance(n, str)}')
else:
    print('no está en ninguna')
```
